## Definicja drzewa decyzyjnego
>***Def.*** Model drzewa decyzyjnego jest algorytmem machine learningowym polegającym na hierarchicznym podziale (segmentacji) danych wejściowych. Model ten można też zdefiniować jako zestaw reguł decyzyjnych prowadzących do ustalonego wyniku.
<p align="center">
  <img src="figures/definicja_drzew_wizualizacja.png" width="600"/>
</p>

### Elementy drzewa decyzyjnego

> 1. Korzeń (root) - miejsce początkowe. Zawiera pełny zbiór danych treningowych.
> 2. Węzeł (Internal node) - segment powstały przez podział danych.
> 3. Liście - ostateczny podział - wynik modelu dla danego zestawu reguł.
> 4. Gałąź - Węzeł wraz z jego kolejnymi podziałami.

> 1. Może służyć do regresji oraz klasyfikacji.
> 2. Prosty w interpretacji (można go łatwo zwizualizować).
> 3. Łatwo użyć w nim zmiennych ilościowych oraz jakościowych.

## Działanie algorytmu
> 1. Zaczynamy od pierwszego węzła (root node).
> 2. Wybieramy cechę i wartości które utworzą gałęzie.
> 3. Uzyskujemy podzbiory.
> 4. Staramy się aby kolejne podzbiory były jak najczystsze, czyli zawierały jak najwięcej elementów z jednej wartości cechy.
> 5. Wyznaczamy kolejne węzły i podziały na kolejne gałęzie aż do uzyskania możliwie "czystych" zbiorów.
> 6. Dla modelu klasyfikacyjnego predykcje mogą zawierać prawdopodobieństwo (na podstawie odsetka przypadków w zbiorze treningowym) lub przynależność do klasy.
> 7. Dla modelu regresyjnego wynikiem będzie wartość średnia w grupie.

### Dodatkowe uwagi
> 1. Algorytm jest chciwy ( greedy ), ponieważ w każdym kroku wybiera najlepszą opcję na stan znany obecnie (nie analizuje przyszłości).
> 2. Podziały dokonywane są poprzez zminimalizowanie MSE, MAE - model regresyjny, natomiast dla drzew klasyfikacyjnych Gini
index lub cross entropy.


### Kryteria nieczystości
| GINI | Entropia
| ----------- | ----------- |
| !["gini"](figures/gini.JPG)       | !["ent"](figures/entropia.JPG)  |


> - Gini Index oraz entropia nazywane są kryteriami nieczystości, gdyż ich wartości pokazują jak bardzo podzbiór jest zanieczyszczony w związku z występowaniem różnych wartości zmiennej objaśnianej.
> - Dążymy do minimalizacji nieczystości przy kolejnych podziałach
> - Dla podzbioru zawierającego tylko pozytywne lub negatywne decyzje entropia i Gini wynosi 0 i oznacza 100% pewność
> - Dla zbioru 'pół na pół', entropia wynosi 1 a Gini 0.5 i oznacza to 50% pewności
> - Oba kryteria zwykle dają ten sam rezultat ale nie zawsze tak musi być.

Zadanie dodatkowe: Stwórz funkcję wyznaczającą indeks giniego dla klasyfikacji binarnej.

```python
def entropia(y: int, n:int):
    t = y+n
    return -y/t*math.log(y/t,2) -n/t*math.log(n/t,2)
```

```python
entropia(9,5)
```

```python
entropia(6,2)
```

```python
def gini_index(y: int, n:int):
    t = y+n
    return 1 -(y/t)**2 -(n/t)**2
```

```python
gini_index(9,5)
```

## Wady i zalety drzew decyzyjnych
### Wady 
#### Overfitting
>* Nasz model powinien być jednocześnie jak najbardziej ***szczegółowy*** i jak najbardziej ***uogólniony***.
>* ***Szczegółowy*** - dopasowany do naszych danych, do naszego projektu, 
>* ***Uogólniony*** - nie może być zbyt dopasowany do danych, które widział, bo nam najbardziej chodzi o przewidywanie wartości nieznanych
>* Drzewa dążą do tego by rozrastać się aż do uzyskania czystych zbiorów w liściach. Często wiąże się to z tym, że w praktyce takie drzewo w zasadzie “zapamiętuje” zbiór treningowy. 
>* Szczególnie jest to widoczne w przypadku regresji. 
>* W praktyce uzyskanie 100% czystości prowadzi do przeuczenia i powoduje, że w każdym liściu jest zbyt mała liczba przypadków.
>* Aby zredukować efekty overfittingu możemy ustawić odpowiednie hiperparametry modelu. 
<p align="center">
  <img src="figures/overfitting.JPG" />
</p>

### Wady - podsumowanie
> 1. Łatwość przeuczenia i niestabilność.
> 2. Niska jakość predykcji w porównaniu do bardziej zaawansowanych technik.
> 3. Potrzeba wybrania pomiędzy jakością predykcji a stabilnością.

### Zalety
> 1. Łatwość interpretacji.
> 2. Znajdowanie zależności nieliniowych.
> 3. Możliwość łatwego wykorzystania zmiennych jakościowych.
> 4. Automatyczny dobór zmiennych istotnych.
> 5. Odporne na braki danych.
> 6. Brak konieczności standaryzacji zmiennych.

## Hiperparametry
>W modelu drzewa deycyjnego będziemy ustalać hiperparametry, służące do stworzenia lepszego, bardziej stabilnego rozwiązania lub wpływających na postać modelu.
>
> 1. max_depth - głębokość drzewa.
> 2. min_samples_leaf - minimalna liczba obserwacji w liściu.
> 3. max_leaf_nodes - maksymalna liczba liści w drzewie.
> 4. max_features - maksymalna liczba zmiennych w modelu.
> 5. ccp_alpha - jak mocno przycinane są drzewa.
> 6. criterion - wybór kryterium ustalającego podział (gini, entropy, mse, mae).
> 7. min_samples_split - minimalna liczba obserwacji do rozważania kolejnego podziału.
