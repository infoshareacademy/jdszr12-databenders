```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("AviationData2.csv", header=None, na_values='?')
df.head()
```

```python
for i in range(df.shape[1]):
# count number of rows with missing values
    n_miss = df[[i]].isnull().sum()
    perc = n_miss / df.shape[0] * 100
    print('> %d, Missing: %d (%.1f%%)' % (i, n_miss, perc))
```

```python
# Method 1: Using column indexing
columns_to_remove = ['Event.Id', 'Accident.Number', 'Latitude', 'Longitude', 'Registration.Number', 'Unnamed: 31']

# Method 2: Using drop() function
df = df.drop(columns=columns_to_remove, inplace=True)
```
